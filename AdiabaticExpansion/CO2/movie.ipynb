```python
import gromacs2
import pathlib
import numpy as np

name = "L"
cell = np.diag([46.37, 46.37, 46.37 * 5]) / 10  # nm

name = "sbl"
cell = np.diag([53.09, 53.09, 53.09 * 5]) / 10  # nm

outfile = open(f"{name}.gro", "w")

root_directory = pathlib.Path(name)
for path_object in sorted(root_directory.rglob("fort.*")):
    if path_object.is_file():
        print(path_object)
        with open(path_object) as f:
            atoms = []
            f.readline()
            for line in f.readlines():
                cols = line.split()[1:]
                atoms.append([float(x) for x in cols])
            atoms = np.array(atoms).reshape(-1, 3) / 10  # nm
            frame = gromacs2.Frame(
                position=atoms,
                atom_name=["C", "O", "O"] * (atoms.shape[0] // 3),
                atom_id=np.arange(atoms.shape[0]) + 1,
                cell=cell,
                residue_name=["CO2"] * atoms.shape[0],
                residue_id=np.arange(atoms.shape[0]) // 3 + 1,
            )
            frame.write_gro(outfile)
```

温度の表示。


```python
import matplotlib.pyplot as plt

temp = []
with open(f"{name}/temp.d") as f:
    temp = [[float(x) for x in line.split()] for line in f.readlines()]

temp = np.array(temp)
temp
```

```python
from IPython.display import HTML
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib.patches import Rectangle

fig = plt.figure(figsize=(5, 2), layout="tight")
ax = fig.add_subplot(111)
# fig.set_size_inches(5, 3)

timerange = temp[-1, 0]


def animate(i):
    ax.clear()
    # Get the point from the points list at index i
    ax.plot(temp[:, 0], temp[:, 1])
    # Plot that point using the x and y coordinates
    ax.plot(temp[i, 0], temp[i, 1], color="red", label="original", marker="o")
    # Set the x and y axis to display a fixed range
    ax.add_patch(
        Rectangle(
            (temp[i, 0], 150), timerange - temp[i, 0], 280, alpha=0.5, facecolor="gray"
        )
    )
    ax.set_xlim([0, timerange])
    ax.set_ylim([200, 280])
    ax.set_xlabel("Time, ps")
    ax.set_ylabel("Temperature, K")


ani = FuncAnimation(fig, animate, frames=len(temp), interval=40, repeat=False)
plt.close()
```

```python
# Show the animation
HTML(ani.to_html5_video())
ani.save(f"{name}.temp.mp4")
```
